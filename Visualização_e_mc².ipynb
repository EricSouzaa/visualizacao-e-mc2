# üí° Visualiza√ß√£o de E = mc¬≤ com Matplotlib

Este mini projeto visualiza a famosa equa√ß√£o de Einstein, `E = mc¬≤`, usando dois gr√°ficos:

- Um gr√°fico em **escala linear** (normal)
- Um gr√°fico com **escala logar√≠tmica no eixo Y**

A ideia √© mostrar como a energia cresce com a massa e como a escala log ajuda a visualizar melhor esse crescimento.


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = np.linspace(0, 10, 11)  # massa de 0 a 10 g
c = 3 * (10**8)             # velocidade da luz em m/s
e = m * (c**2)              # E = mc¬≤
```

## üìà Cria√ß√£o dos gr√°ficos

Aqui vamos visualizar dois gr√°ficos lado a lado:

- √Ä esquerda: escala normal
- √Ä direita: escala logar√≠tmica no eixo Y


```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Gr√°fico com escala normal
axes[0].plot(m, e, color='blue')
axes[0].set_xlabel('Mass (g)')
axes[0].set_ylabel('Energy (J)')
axes[0].set_title('E = mc¬≤ (Escala Normal)')
axes[0].grid(True)

# Gr√°fico com escala logar√≠tmica no eixo Y
axes[1].plot(m, e, color='red')
axes[1].set_xlabel('Mass (g)')
axes[1].set_ylabel('Energy (J)')
axes[1].set_title('E = mc¬≤ (Escala Logar√≠tmica)')
axes[1].set_yscale('log')
axes[1].grid(True, which='both', axis='y')

plt.tight_layout()
plt.show()
```

## ‚úÖ Conclus√£o

- A equa√ß√£o `E = mc¬≤` mostra que **pequenas massas geram enormes quantidades de energia**, por conta do valor absurdo de `c¬≤`.
- A escala logar√≠tmica ajuda a visualizar o crescimento exponencial da energia √† medida que a massa aumenta.

Esse tipo de visualiza√ß√£o √© √∫til tanto para ensino quanto para comunicar conceitos de f√≠sica de forma visual.
